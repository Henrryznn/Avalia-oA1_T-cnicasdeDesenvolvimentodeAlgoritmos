```python
"Listas"

# Programa para cadastrar nomes de alunos e exibir ao final

alunos = []  # Lista para armazenar os nomes

while True:
    nome = input("Digite o nome do aluno (ou 'sair' para finalizar): ")

    if nome.lower() == "sair":  # Condição de parada
        break

    alunos.append(nome)  # Armazena o nome na lista

alunos.sort() # Adiciona esta linha para ordenar a lista em ordem alfabética

print("\nLista de alunos cadastrados:")
for aluno in alunos:
    print(aluno)
```
