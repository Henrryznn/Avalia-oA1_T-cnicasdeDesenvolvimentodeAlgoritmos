```python
"Estruturas de Repetição"

# Contador de 1 a 100 usando FOR
print("Contador com FOR (números pares de 1 a 100):")
for num in range(1, 101):
    if num % 2 == 0:  # Verifica se o número é par
        print(num)

print("\n---------------------------\n")

# Contador de 1 a 100 usando WHILE
print("Contador com WHILE (números pares de 1 a 100):")
num = 1
while num <= 100:
    if num % 2 == 0:  # Verifica se o número é par
        print(num)
    num += 1  # Incremento
```
