```python
"Dicionários"

# Sistema simples de cadastro de produtos usando dicionário

produtos = {}  # Dicionário para armazenar produtos e preços

while True:
    nome = input("Digite o nome do produto (ou 'sair' para finalizar): ")

    if nome.lower() == "sair":
        break

    # Substitui a vírgula por ponto antes de converter para float
    preco_str = input(f"Digite o preço do produto '{nome}': ")
    preco = float(preco_str.replace(',', '.'))

    produtos[nome] = preco  # Armazena no dicionário (chave: nome, valor: preço)

print("\nLista de produtos cadastrados:")
for nome, preco in produtos.items():
    print(f"Produto: {nome} — Preço: R$ {preco:.2f}")
```
